```python
def ou_process(ml,gl,sl2,mq,gq,sq2,b):
    time=100
    sym = int(1e6)
    np.random.seed(3)
    L = gge.ornstein_uhlenbeck(ml,gl,sl2,shape=(sym,time))
    X = np.cumsum(L,axis=1)
    np.random.seed(5)
    Q = gge.ornstein_uhlenbeck(mq,gq,sq2,shape=(sym,time))
    G=[np.zeros(sym)]
    for k in range(time):
        G.append((1-b)*G[k]+np.exp(X[:,k])*Q[:,k])
    return L,X,Q,np.vstack(G).T
```

```python
def mean_cov_time(t,L,X,Q,G):
    X=X[:,t];L=L[:,t];Q=Q[:,t];G=G[:,t]
    cvlx = np.cov(L,X);cvlq = np.cov(L,Q)
    cvlg = np.cov(L,G);cvxq = np.cov(X,Q)
    cvxg = np.cov(X,G);cvgq = np.cov(G,Q)
    C = np.zeros((4,4))
    C[0,0] = cvxg[0,0];C[0,1]=cvxg[0,1];C[0,2]=cvlx[0,1];C[0,3]=cvxq[0,1]
    C[1,1] = cvxg[1,1]; C[1,2] = cvlg[0,1]; C[1,3]=cvgq[0,1]
    C[2,2] = cvlq[0,0];C[2,3]=cvlq[0,1];C[3,3]=cvlq[1,1]
    C[1,0]=C[0,1];C[2,0]=C[0,2];C[3,0]=C[0,3];C[2,1]=C[1,2];C[3,1]=C[1,3]
    C[3,2]=C[2,3]
    return np.array([np.mean(X),np.mean(G),np.mean(L),np.mean(Q)]).reshape((4,1)),C
```

```python
ml=1e-02;gl=1/100;sl2=1e-6;mq=0.5;gq=1/50;sq2=5;b=1e-05
L,X,Q,G = ou_process(ml,gl,sl2,mq,gq,sq2,b)
```
